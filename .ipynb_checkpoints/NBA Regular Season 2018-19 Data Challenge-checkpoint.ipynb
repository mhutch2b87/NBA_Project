# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
teamWS = nbaStat[['Team', 'Name', 'WS']]
sorted_teams = teamWS.sort_values('Team')


nbaStat.groupby('Team').sum()[['WS']].nlargest(1,columns='WS')
```

# What player had the best 3-pt percentage? 

```python
#FG3%
nbaStat[['Name', 'FG3%']].sort_values('FG3%', ascending=False)
```

# Who played the most minutes during the season

```python
nbaStat.groupby('Name').sum()[['MP']].nlargest(1,columns='MP')
```

# What player given their player effiecency rating was the clutchest during the season?

```python
nbaStat.groupby('Name').median()[['PER']].nlargest(1,columns='PER')
```

# What team had the youngest roster?

```python
nbaStat.groupby('Team').sum()[['Age']].sort_values('Ages')
```

# Who is the highest paid player during the seasion?

```python
nbaStat[['Name', 'Salary']].sort_values('Salary', ascending = False)
# type(nbaStat['Salary'])
# nbaStat['Salary'].nlargest(1,'Salary')
```
